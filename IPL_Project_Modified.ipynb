# DATA ANALYSIS ON IPL DATASET

### This project is the data analysis and data exploration on IPL Dataset to know the insights of the IPL Data. 
#### Libraries used in this project:
* Numpy
* Pandas
* Matplotlib
* Seaborn

#### <u>There are 18 features and 577 records in the dataset</u>
#### Attributes of the Dataset are:
* <b>id</b> - the unique ID of any particular match
* <B>season</B> - year or session in which the match was held
* <b>city</b> - city where match was conducted
* <b>date</b> - date of the match
* <b>team1</b> - team played the match
* <b>team2</b> - another team which played the match
* <b>toss_winner</b> - team who won the toss in any particular match
* <b>toss_decision</b> - decision took by team after winning the toss i.e. bat or field
* <b>result</b> - result of the match i.e. tie or normal
* <b>dl_applied</b> - DLS (Duckworth - Lewis - Stern) Method applied or not
* <b>winner</b> - winning team of the match
* <b>win_by_run</b> - runs by which team won the match
* <b>win_by_wickets</b> - wickets by which team won the match
* <b>player_of_match</b> - Man or Player of the match
* <b>venue</b> - Stadium where match was conducted
* <b>umpire1</b> - 1st umpire on the field
* <b>umpire2</b> - 2nd umpire on the field
* <b>umpire3</b> - 3rd umpire of the match
##### The Duckworth–Lewis–Stern method (DLS) is a mathematical formulation designed to calculate the target score for the team batting second in a limited overs cricket match interrupted by weather or other circumstances.

## 11.	Print the team names who won the match with the double digit and both are the same.(Ex: 66,33)

```python
def count_digit(n):
    count=0
    copy=n
    response=None
    while copy>0:
        copy=copy//10
        count+=1
    if count==2:
        x=n//10
        rem=n%10
        if x==rem:
            response=n
    return response
```

```python
temp = df[['winner','win_by_runs']]
```

```python
temp["win_by_runs"]=temp['win_by_runs'].apply(count_digit)
temp= temp.dropna()
temp
```

## 18.	Print umpire names who did the umpiring for maximum and minimum times

```python
ser3=df['umpire2'].value_counts()
ser4=df['umpire1'].value_counts()
df3=pd.concat([ser3,ser4],axis=1,join='outer')
df3.fillna(0,inplace=True)
df3['sum']=df3['umpire1']+df3['umpire2']
print("Umpire with maximum number of times:")
print(df3[df3['sum']==max(df3['sum'])]['sum'])
print("\n\nUmpire with minimum number of times:")
print(df3[df3['sum']==min(df3['sum'])]['sum'])
```
